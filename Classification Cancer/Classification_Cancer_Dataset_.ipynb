```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

df['diagnosis'] = LE.fit_transform(df['diagnosis'])
```

```python
# Decide Input / Output Features.

X = df.drop(['id', 'diagnosis', 'Unnamed: 32'], axis = 1 )                  # Input
Y = df['diagnosis']
```

```python
# Split Data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.20)
```

```python
# Apply Logistic Regression , Decision Tree

from sklearn.linear_model import LogisticRegression
from sklearn.tree  import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# 1. Logistic Regression
```

```python
LR = LogisticRegression(max_iter= 10000)
```

```python
# Train Model
LR.fit(X_train, Y_train)
```

```python
print(" Logistic Regression : ")
print(" Accuracy \t : ", accuracy_score(Y_test , LR.predict(X_test)   ) )
```

```python
# 2. Decision Tree
```

```python
DT = DecisionTreeClassifier()
```

```python
# Train Model
DT.fit(X_train, Y_train)
```

```python
print(" Decision Tree : ")
print(" Accuracy \t : ", accuracy_score(Y_test , DT.predict(X_test)   ))
```
